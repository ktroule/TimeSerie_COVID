# Exploratory analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Load data

```python
covid = pd.read_csv('../data/processed/time_serie_2nd_orderdiff.csv',
    index_col = 'Date',
    parse_dates = True)

# Conver to serie
covid = pd.Series(covid['0'], index = covid.index)
covid.head()
```

### Split train/test set
The data extent of the test set is related to the extent of the forecast. In this exercise we aim to forecast 1 month, thus any data prio 2022-04-15 will be used as training and the remaing month as test set.

```python
cutoff = '2022-04-15'

train = covid[covid.index <= cutoff]
test = covid[covid.index > cutoff]

# Plot train/test
sns.lineplot(train)
sns.lineplot(test)
```
